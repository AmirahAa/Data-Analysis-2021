Worked example 1: 

Suppose a dice is rolled 5 times. What is the probability of getting exactly 2 fours?

```python
# number of successes 
nu = 2
# no of trials
N= 100
#prob 
theta = 1./6

from scipy.stats import binom
import math

prob_cheat = binom.pmf(nu,N,theta)

print(prob_cheat)

prob = theta**nu *(1-theta)**(N-nu)*math.factorial(N)/(math.factorial(N-nu)*math.factorial(nu))
print(prob)

x = np.arange(0,100,1)
#x = np.arange(binom.ppf(0.01, N, theta),binom.ppf(0.99, N, theta))

print(x)

plt.plot(x, binom.pmf(x, N, theta), 'bo', ms=8, label='binom pmf')
```
