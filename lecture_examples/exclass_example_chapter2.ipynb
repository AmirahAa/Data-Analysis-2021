```python
import numpy as np
import pylab as plt
%matplotlib inline
```

Worked example 1: 

Suppose a dice is rolled 5 times. What is the probability of getting exactly 2 fours?

```python
# number of successes 
nu = 2
# no of trials
N= 5
#prob 
theta = 1./6

from scipy.stats import binom
import math

prob_cheat = binom.pmf(nu,N,theta)

print(prob_cheat)

prob = theta**nu *(1-theta)**(N-nu)*math.factorial(N)/(math.factorial(N-nu)*math.factorial(nu))
print(prob)
```

Worked example 2: 

We are given a set of student grades for a single exam and we find that it is best fit by a Beta distribution with a = 8.28 and b = 3.16.

What is the probability a student achieves less than the mean grade?

```python
from scipy.stats import beta
a = 8.28
b=3.16

mean = a/(a+b)
print(mean)
```

```python
prob_beta = beta(a,b)

x = np.linspace(0,1)

plt.plot(x,prob_beta.pdf(x))
plt.axvline(mean,c='red')
```

```python
plt.plot(x,beta.cdf(x,a,b))
plt.axhline(mean,c='red')
```

```python
prob = beta.cdf(mean,a,b)
print(prob)
```
